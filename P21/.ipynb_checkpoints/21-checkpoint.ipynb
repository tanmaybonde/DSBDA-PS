```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("adult.csv", names=["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"])
```

```python
col = ['age', 'education-num', 'hours-per-week']
for c in col:
    df[c].hist(figsize= (10, 6))
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['fnlwgt'], 'bo')
plt.show()
```

```python
col=['age','education-num','hours-per-week']

for c in col:
    plt.figure(figsize=(10, 6))
    df[c].value_counts().plot(kind='bar')
    plt.show()
```

```python
plt.figure(figsize=(18, 3))
plt.plot(df['hours-per-week'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='education-num', y ='hours-per-week', hue='income')
plt.show()
```
