```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("adult.csv", names=["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"])
```

```python
df
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.replace("?", 0, inplace=True)
```

```python
outliers = np.abs((df - df.mean()) / df.std()) > 3
```

```python
df = df[~outliers.any(axis=1)]
```

```python
df
```

```python
labelencoder = LabelEncoder()
cols = ["workclass", "education", "marital-status", "occupation", "relationship", "race", "sex", "native-country", "income"]

for col in cols:
    df[col] = labelencoder.fit_transform(df[col])
```

```python
df
```

```python
X = df.drop("income", axis=1)
Y = df['income']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=45)
```

```python
modelLog = LogisticRegression()
modelLog.fit(X_train, Y_train)
```

```python
y_pred = modelLog.predict(X_test)
```

```python
accuracy_score(Y_test, y_pred)
```

```python
modelG = GaussianNB()
modelG.fit(X_train, Y_train)
```

```python
y_pred = modelG.predict(X_test)
```

```python
accuracy_score(Y_test, y_pred)
```

```python
modelM = MultinomialNB()
modelM.fit(X_train, Y_train)
```

```python
y_pred = modelM.predict(X_test)
```

```python
accuracy_score(Y_test, y_pred)
```
