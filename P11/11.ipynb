```python
movieDf.rename(columns={'id': 'movieId'}, inplace=True)
```

```python
movieDf['movieId'] = pd.to_numeric(movieDf['movieId'], errors='coerce')
```

```python
meanRatingDf = ratingDf.groupby(by = 'movieId')['rating'].mean()
```

```python
movieDf = movieDf.join(other=meanRatingDf, on='movieId')
```

```python
movieDf.sort_values(by = 'rating', inplace= True)
```
