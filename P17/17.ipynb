```python
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
df = pd.read_csv("hung.csv", names=["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal","target"])
```

```python
df
```

```python
df.isna().sum()
```

```python
df.replace("?", 0, inplace=True)
df = df.astype(int)
```

```python
col=['age','chol','thalach']
df[col].hist(bins=10,figsize=(10,6))
plot.tight_layout()
plot.show()
```

```python
col = ['sex', 'cp', 'target']

for c in col:
	sns.catplot(data=df, x=c, kind='count', aspect=1.5)

plot.plot()
```

```python
df['restecg'].value_counts().plot(kind='bar')
plot.xlabel = c
plot.ylabel = 'count'
    
plot.plot()
```

```python
df['fbs'].value_counts().plot(kind='bar')
plot.xlabel = c
plot.ylabel = 'count'
    
plot.plot()
```

```python
df['slope'].value_counts().plot(kind='bar')
plot.xlabel = c
plot.ylabel = 'count'
    
plot.plot()
```

```python
columns = ['trestbps', 'oldpeak', 'thalach']
for column in columns:
    plot.plot(df[column])
    plot.show()
```

```python
sns.scatterplot(data=df, x='chol', y='thalach', hue='target')
plot.xlabel = 'chol'
plot.ylabel = 'thalach'
plot.show()
```
