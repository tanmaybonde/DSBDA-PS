```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
```

```python
request=requests.get("https://www.amazon.in/ASUS-Vivobook-Integrated-Graphics-M1603QA-MB502WS/dp/B09ZV5CDWD/ref=asc_df_B09ZV5CDWD/?tag=googleshopdes-21&linkCode=df0&hvadid=586198302882&hvpos=&hvnetw=g&hvrand=6163846051670449254&hvpone=&hvptwo=&hvqmt=&hvdev=c&hvdvcmdl=&hvlocint=&hvlocphy=9301538&hvtargid=pla-1678103515170&th=1")
request
```

```python
soup=bs(request.content,"html.parser")
```

```python
print(soup.prettify())
```

```python
reviews=soup.findAll('a',attrs={'data-hook':'review-title'});
for review in reviews:
    print(review.get_text())
```

```python
review_content = []
for i in range(0,len(reviews)):
    review_content.append(reviews[i].get_text())
review_content
```

```python
comments=soup.findAll('div',{'data-hook':'review-collapsed'});
for comment in comments:
    print(comment.get_text()+"\n\n")
```

```python
comment_content=[]
for i in range(0,len(comments)):
    comment_content.append(comments[i].get_text())
comment_content    
```

```python
cust_name=soup.find_all('span',{'class':'a-profile-name'});
for names in cust_name:
    print(names.get_text()+"\n\n")
```

```python
customer=[]
for i in range(0,len(cust_name)):
    customer.append(cust_name[i].get_text())
customer    
```

```python
ratings=soup.find_all('span',{'class':'a-icon-alt'});
for rating in ratings:
   print(rating.get_text()+"\n\n")
```

```python
rate=[]
for i in range(0,len(ratings)):
    rate.append(ratings[i].get_text())
rate    
```

```python
# data={
#     "Name":cust_name,
#     "Review":reviews,
#     "Comment":comments,
#     "Rating":ratings
# }
df=pd.DataFrame()
df['Customer Name']=customer[:5]
df['Review']=review_content[:5]
df['Comment']=comment_content[:5]
df['Rating']=rate[:5]
```
