```python
df = pd.read_csv("adult.csv", names=["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"])
```

```python
contries = df['native-country'].unique()
contrySubsets = {}

for contry in contries:
    subsets[contry] = df[df['native-country'] == contry]
```

```python
allSex = df['sex'].unique()
sexSubsets = {}

for sex in allSex:
    sexSubsets[sex] = df[df['sex'] == sex]
```

```python
allRace = df['race'].unique()
raceSubsets = {}

for race in allRace:
    raceSubsets[race] = df[df['race'] == race]
```
